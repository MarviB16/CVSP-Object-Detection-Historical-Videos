```python
import os, cv2
from tqdm.autonotebook import tqdm
from xml.dom import minidom

path = '/Users/marvin_burges/Documents/ComputerVisionSystemProgramming/OpenLabeling/main/input_rum/'
path_xml = '/Users/marvin_burges/Documents/ComputerVisionSystemProgramming/OpenLabeling/main/output_rum/PASCAL_VOC/'

paths = os.listdir(path)
for x in tqdm(range(len(paths))):
    y = paths[x]
    mydoc = minidom.parse(path_xml+y[:-4] + ".xml")
    items = mydoc.getElementsByTagName('object')
    
    if len(items) > 0:
        image = cv2.imread(path+y)
        cv2.imwrite("/Users/marvin_burges/Documents/ComputerVisionSystemProgramming/OpenLabeling/main/input_rum_jpg/" + y[:-4] + ".jpg", image)
```
